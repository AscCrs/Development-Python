### Graficación de una serie geometrica de Fourier

Se busca expandir y gráficar la siguiente serie de Fourier 

$f(t) = \left(\frac{16}{\pi^2} \sum_{n; impar}^{\infty}\frac{1}{n^2}\cos{n}w_{0}{t} \right)$

$w_{0} = \frac{2\pi}{T}$

En 3 momentos de n distintos
n = 10
n = 100
n = 1000

Teniendo que el periodo vale 5s 

T = 5

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Expansión de la serie de Fourier
$f(t) = \left(\frac{16}{\pi^2} \sum_{n; impar}^{\infty}\frac{1}{n^2}\cos{n}w_{0}{t} \right)$

```python
def fourier(t, n):
    w0 = 2 * np.pi / 5
    suma = 0
    for i in range(1, n+1, 2):
        suma += 1 / (i ** 2) * np.cos(i * w0 * t)
    return suma * (16 / np.pi ** 2)
```

```python
# Parámetros
Tmin = -5/2
Tmax = 5/2
T = Tmax - Tmin
n1 = 10
n2 = 100
n3 = 1000
```

```python
# Graficar la expansión de Fourier
t = np.linspace(0, 10, 500)
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.plot(t, fourier(t, n1), lw=1, label=f'Aproximación con n={n1}')
ax.plot(t, fourier(t, n2), lw=1, label=f'Aproximación con n={n2}')
ax.plot(t, fourier(t, n3), lw=1, label=f'Aproximación con n={n3}')
```

```python
ax.legend(loc='best')
ax.set_xlabel('t')
ax.set_ylabel('f(t)')
ax.set_title('Expansión de Fourier para f(t)')
plt.show()
```
